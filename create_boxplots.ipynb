```
import seaborn as sns
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```
scores_BLEU = pd.DataFrame(columns=["data", "model", "score"])
scores_ROUGE = pd.DataFrame(columns=["data", "model", "score"])
scores_BERT = pd.DataFrame(columns=["data", "model", "score"])

# hp scores GRU
with open('/content/gdrive/My Drive/TxMM/GRU_results/hp_scores_BLEU.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BLEU = scores_BLEU.append({'data': 'fiction', 'model': 'GRU', 'score': data[str(i)]}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GRU_results/hp_scores_ROUGE.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_ROUGE = scores_ROUGE.append({'data': 'fiction', 'model': 'GRU', 'score': data[str(i)]['R']}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GRU_results/hp_scores_BERT.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BERT = scores_BERT.append({'data': 'fiction', 'model': 'GRU', 'score': data[str(i)]}, ignore_index=True)


# hp scores GPT-2
with open('/content/gdrive/My Drive/TxMM/GPT-2_results/hp_scores_BLEU.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BLEU = scores_BLEU.append({'data': 'fiction', 'model': 'GPT-2', 'score': data[str(i)]}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GPT-2_results/hp_scores_ROUGE.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_ROUGE = scores_ROUGE.append({'data': 'fiction','model': 'GPT-2', 'score': data[str(i)]['R']}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GPT-2_results/hp_scores_BERT.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BERT = scores_BERT.append({'data': 'fiction','model': 'GPT-2', 'score': data[str(i)]}, ignore_index=True)


# news scores GRU
with open('/content/gdrive/My Drive/TxMM/GRU_results/news_scores_BLEU.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BLEU = scores_BLEU.append({'data': 'non-fiction', 'model': 'GRU', 'score': data[str(i)]}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GRU_results/news_scores_ROUGE.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_ROUGE = scores_ROUGE.append({'data': 'non-fiction', 'model': 'GRU', 'score': data[str(i)]['R']}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GRU_results/news_scores_BERT.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BERT = scores_BERT.append({'data': 'non-fiction', 'model': 'GRU', 'score': data[str(i)]}, ignore_index=True)


# news score GPT-2
with open('/content/gdrive/My Drive/TxMM/GPT-2_results/news_scores_BLEU.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BLEU = scores_BLEU.append({'data': 'non-fiction', 'model': 'GPT-2', 'score': data[str(i)]}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GPT-2_results/news_scores_ROUGE.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_ROUGE = scores_ROUGE.append({'data': 'non-fiction','model': 'GPT-2', 'score': data[str(i)]['R']}, ignore_index=True)

with open('/content/gdrive/My Drive/TxMM/GPT-2_results/news_scores_BERT.json') as json_file:
    data = json.load(json_file)
    for i in range(0,100):
      scores_BERT = scores_BERT.append({'data': 'non-fiction','model': 'GPT-2', 'score': data[str(i)]}, ignore_index=True)     
```

# Box plots for GRU ROUGE BLEU and BERT scores

```
# BLEU
sns.set_style("whitegrid") 
sns_box = sns.boxplot(x = 'model', y = 'score', hue='data', data=scores_BLEU).set_title("BLEU scores")
plt.show()
fig = sns_box.get_figure()
fig.savefig('/content/gdrive/My Drive/TxMM/figures/boxplot_BLEU.jpg') 

# ROUGE
sns.set_style("whitegrid") 
sns_box = sns.boxplot(x = 'model', y = 'score', hue='data', data=scores_ROUGE).set_title("ROUGE scores")
plt.show()
fig = sns_box.get_figure()
fig.savefig('/content/gdrive/My Drive/TxMM/figures/boxplot_ROUGE.jpg') 

# BERT
sns.set_style("whitegrid") 
sns_box = sns.boxplot(x = 'model', y = 'score', hue='data', data=scores_BERT).set_title("BERTCoSim scores")
plt.show()
fig = sns_box.get_figure()
fig.savefig('/content/gdrive/My Drive/TxMM/figures/boxplot_BERT.jpg') 
```
